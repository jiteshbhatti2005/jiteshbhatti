```python
########################################################################################################
#Name : - Jitesh Bhatti 
#Enrollment :- 216270307057
#Subject :- IML
#Sem :- 5
########################################################################################################

#Practical - 3 : Write a NumPy program to implement following operation

#Practical - 3.1 to stack arrays horizontally (column wise)

import numpy as np

arr1 = np.arange(1,6)
print("Array 1 : ",arr1)

arr2 = np.arange(6,11)
print("Array 2 : ",arr2)

arr3 = np.hstack((arr1,arr2))
print("\nColoum wise : arr3\n",arr3)


arr4 = np.vstack((arr1,arr2))
print("\nRow wise : \n",arr4)
```
