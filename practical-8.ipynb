```python
#######################################################################################################
#Name : - Jitesh Bhatti 
#Enrollment :- 216270307057
#Subject :- IML
#Sem :- 5
#######################################################################################################

#Practical - 8 : Write a Pandas program to create a line plot of the opening, closing stock
#                prices of given company between two specific dates.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

start_date = '2008-11-27'
end_date = '2009-02-06'

filtered_data = df.loc[start_date:end_date]

plt.figure(figsize=(10, 6))
plt.plot(filtered_data.index, filtered_data['Open'], label='Opening Price', color='blue')
plt.plot(filtered_data.index, filtered_data['Close'], label='Closing Price', color='green')

plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Prices between {} and {}'.format(start_date, end_date))

plt.legend()

plt.show()
```
